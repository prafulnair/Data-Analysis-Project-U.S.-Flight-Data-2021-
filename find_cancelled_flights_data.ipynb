```python
import pandas as pd
import math

# TO FIND THE Flights that have been cancelled between start_date = '2021-09-01' and end_date = '2021-09-30'
```

```python
df = pd.read_csv('Dataset/Combined_Flights_2021.csv')
df
```

```python
start_date = '2021-09-01'
end_date = '2021-09-30'
# Select rows only for September month
mask = (df['FlightDate'] >= start_date) & (df['FlightDate'] <= end_date)
df2 = df.loc[mask]
print(df2)
```

```python
df2 = df2.loc[(df2['Cancelled'] == True)]
df2
```

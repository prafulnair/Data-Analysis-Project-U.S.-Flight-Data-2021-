```python
import pandas as pd
import math
## Getting Mean Airtime for flights between BNA and ORD
```

```python
df = pd.read_csv('Dataset/Combined_Flights_2021.csv')
df
```

```python
list(df.columns)
```

```python
mask = (df['Origin'] == "BNA") & (df['Dest'] == "ORD")
df2 = df.loc[mask]
print(df2)
```

```python
mean_airtime = df2['AirTime'].mean()
formatted_airtime = f"{mean_airtime:.2f}"
formatted_airtime
```
